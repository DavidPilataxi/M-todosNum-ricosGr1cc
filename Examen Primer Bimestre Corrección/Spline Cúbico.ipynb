```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el spline c√∫bico manualmente
def spline_cubico(t):
    # Coeficientes calculados
    a = 1
    b = 1
    c = -0.5
    d = 0.25
    t0 = -1  # Punto inicial del intervalo

    # Expresi√≥n del spline
    return a + b * (t - t0) + c * (t - t0)**2 + d * (t - t0)**3

# Imprimir la ecuaci√≥n del spline
print("El spline c√∫bico S(t) es:")
print("S(t) = 1 + (t + 1) - 0.5 * (t + 1)^2 + 0.25 * (t + 1)^3")

# Rango de valores para graficar
t_vals = np.linspace(-1, 1, 100)
y_spline = spline_cubico(t_vals)

# Graficar el spline c√∫bico
plt.figure(figsize=(8, 6))
plt.plot(t_vals, y_spline, label='Curva Interpolada', color='purple', linestyle='--')

# Puntos originales
puntos_x = [-1, 1]
puntos_y = [1, 3]
plt.scatter(puntos_x, puntos_y, color='orange', marker='x', label='Datos')

# A√±adir etiquetas y leyenda
plt.title('Interpolaci√≥n con Spline C√∫bico')
plt.xlabel('Eje Horizontal (t)')
plt.ylabel('Valores Interpolados')
plt.grid(True)
plt.legend()
plt.show()
```
