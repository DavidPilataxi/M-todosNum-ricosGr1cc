1. Use el método de bisección para encontrar soluciones precisas dentro de 10^(-2)  para x^3-7x^2+14x-6=0 en cada intervalo:


a.[0,1]

```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo
def f(x):
    return x**3 - 7 * x**2 + 14 * x - 6

# Método de Bisección
def bisec(f, a, b, tol=1e-2, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Intervalo no válido para el método de bisección.")
        return None
    
    for _ in range(max_iter):
        m = (a + b) / 2
        if abs(b - a) / 2 <= tol or f(m) == 0:
            return m  # Raíz aproximada o exacta
        elif f(a) * f(m) < 0:
            b = m
        else:
            a = m
    return (a + b) / 2

# Intervalo y ejecución
a, b = 0, 1
root = bisec(f, a, b)

# Gráfica


x_vals = np.linspace(a, b, 500)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^3 - 7x^2 + 14x - 6", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle="--", label=f"RaIzzz ≈ {root:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Metodo de Biseccion para Aproximacion de Raiz")
plt.legend()
plt.grid(True)
plt.show()
```

4. a. Dibuje las gráficas para y= x^2-1 y y=e^(1-x^2 )

```python
import numpy as np
import matplotlib.pyplot as plt

# Función exponencial
def h(x):
    return x**2 - 1

# Valores de x y y para la gráfica
x_vals = np.linspace(-3, 3, 500)
y_vals = h(x_vals)

# Graficar y = e^(1 - x^2)
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$y = e^{1 - x^2}$", color='green')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafica de y = y = x^2 - 1")
plt.legend()
plt.grid(True)
plt.show()

```
