### Ejercicio 1 a

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
x_vals = [4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1]
y_vals = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72]

# Definición de la función lineal
def linear_func(x, slope, intercept):
    return slope * x + intercept

# Ajuste de parámetros
params, _ = curve_fit(linear_func, x_vals, y_vals)
slope, intercept = params

# Creación de la función ajustada
def fitted_func(x):
    return slope * x + intercept

# Calcular puntos de la línea ajustada
x_interp = np.linspace(3, 8, 100)
y_interp = fitted_func(x_interp)

# Cálculo del error cuadrático
def calculate_squared_error(observed, predicted_func, x_data):
    total_error = 0
    for i in range(len(observed)):
        total_error += (observed[i] - predicted_func(x_data[i])) ** 2
    return total_error

# Visualización
plt.scatter(x_vals, y_vals, color='blue', label="Datos observados")
plt.plot(x_interp, y_interp, color='red', label="Línea ajustada")
plt.title("Ajuste Lineal")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()  

# Resultados
print(f'{round(slope, 3)} x + {round(intercept, 3)}')
print(f'Error cuadrático: {calculate_squared_error(y_vals, fitted_func, x_vals)}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos de entrada
x_data = [4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1]
y_data = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72]

# Definir la función potencial
def power_function(x, coef, exponent):
    return coef * x**exponent

# Ajustar los datos al modelo
params, _ = curve_fit(power_function, x_data, y_data)
coef, exponent = params

# Crear la función ajustada
def fitted_power_function(x):
    return coef * x**exponent

# Generar puntos para la curva ajustada
x_fit = np.linspace(3, 8, 100)
y_fit = fitted_power_function(x_fit)

# Función para calcular el error cuadrático
def calculate_squared_error(observed, func, x_vals):
    errors = [(observed[i] - func(x_vals[i]))**2 for i in range(len(observed))]
    return sum(errors)

# Graficar los datos y el ajuste
plt.scatter(x_data, y_data, color="red", label="Datos originales")
plt.plot(x_fit, y_fit, color="blue", label="Curva ajustada")
plt.title("Ajuste Potencial")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()

# Mostrar resultados
print(f'{round(coef, 3)} * x^{round(exponent, 3)}')
print(f'Error cuadrático: {calculate_squared_error(y_data, fitted_power_function, x_data)}')
```
