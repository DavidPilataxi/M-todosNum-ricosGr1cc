### Gr√°fica de cos(X)

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango para el eje x
x_vals = np.linspace(-8, 8, 400)

# Aproximaciones de Taylor para cos(x) con distintos t√©rminos
approx_t2 = 1 - x_vals**2 / 2                # Aproximaci√≥n con 2 t√©rminos
approx_t4 = approx_t2 + x_vals**4 / 24       # Aproximaci√≥n con 4 t√©rminos
approx_t6 = approx_t4 - x_vals**6 / 720      # Aproximaci√≥n con 6 t√©rminos

# Graficamos cos(x) y sus aproximaciones de Taylor
plt.figure(figsize=(10, 6))
plt.plot(x_vals, np.cos(x_vals), label='cos(x)', color='purple')         # Funci√≥n coseno
plt.plot(x_vals, approx_t2, label='Taylor 2 t√©rminos', color='brown', linestyle='-.')
plt.plot(x_vals, approx_t4, label='Taylor 4 t√©rminos', color='teal', linestyle=':')
plt.plot(x_vals, approx_t6, label='Taylor 6 t√©rminos', color='magenta', linestyle='--')

# Configuraci√≥n de la gr√°fica
plt.xlabel("x"), plt.ylabel("y")
plt.title("Aproximaciones de Taylor para cos(x)")
plt.legend(), plt.grid(True)
plt.xlim(-8, 8), plt.ylim(-5, 5)
plt.show()
```

### 2. GRAFICA DE LA FUNCION $1/(1-X)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x_vals = np.linspace(-0.9, 0.9, 400)  # Ajustamos el rango para evitar valores donde la funci√≥n diverge

# Aproximaciones de Taylor para 1 / (1 - x) con distintos t√©rminos
taylor_1 = 1  # Aproximaci√≥n con 1 t√©rmino
taylor_2 = 1 + x_vals  # Aproximaci√≥n con 2 t√©rminos
taylor_3 = 1 + x_vals + x_vals**2  # Aproximaci√≥n con 3 t√©rminos
taylor_4 = 1 + x_vals + x_vals**2 + x_vals**3  # Aproximaci√≥n con 4 t√©rminos
taylor_5 = 1 + x_vals + x_vals**2 + x_vals**3 + x_vals**4  # Aproximaci√≥n con 5 t√©rminos
taylor_6 = 1 + x_vals + x_vals**2 + x_vals**3 + x_vals**4 + x_vals**5  # Aproximaci√≥n con 6 t√©rminos

# Graficamos la funci√≥n original y cada aproximaci√≥n de Taylor
plt.figure(figsize=(10, 6))
plt.plot(x_vals, 1 / (1 - x_vals), label=r'$\frac{1}{1 - x}$', color='black')  # Funci√≥n original
plt.plot(x_vals, taylor_1 * np.ones_like(x_vals), label='1 t√©rmino: 1', color='magenta', linestyle='--')
plt.plot(x_vals, taylor_2, label='2 t√©rminos: 1 + x', color='teal', linestyle='--')
plt.plot(x_vals, taylor_3, label='3 t√©rminos: 1 + x + x¬≤', color='crimson', linestyle='--')
plt.plot(x_vals, taylor_4, label='4 t√©rminos: 1 + x + x¬≤ + x¬≥', color='indigo', linestyle='--')
plt.plot(x_vals, taylor_5, label='5 t√©rminos: 1 + x + x¬≤ + x¬≥ + x‚Å¥', color='darkorange', linestyle='--')
plt.plot(x_vals, taylor_6, label='6 t√©rminos: 1 + x + x¬≤ + x¬≥ + x‚Å¥ + x‚Åµ', color='limegreen', linestyle='--')

# Configuraci√≥n del gr√°fico
plt.xlabel("x")
plt.ylabel("y")
plt.title("Aproximaciones de Taylor de 1 / (1 - x) con distintos t√©rminos")
plt.legend()
plt.grid(True)
plt.ylim(-1, 5)
plt.show()
```

### 3. GRAFICA DE LA FUNCION $ln(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci√≥n original
def original_function(x):
    return np.log(x)

# Definimos el polinomio de Taylor de tercer grado
def taylor_approx(x):
    return (x - 1) - (x - 1)**2 / 2 + (x - 1)**3 / 3

# Rango de valores de x
x_vals = np.linspace(0.1, 2, 400)
orig_values = original_function(x_vals)
taylor_vals = taylor_approx(x_vals)

# Graficamos
plt.figure(figsize=(10, 6))
plt.plot(x_vals, orig_values, label='f(x) = ln(x)', color='darkgreen')
plt.plot(x_vals, taylor_vals, label='Polinomio de Taylor (grado 3)', color='darkred', linestyle='--')
plt.title('Comparaci√≥n: ln(x) y su Polinomio de Taylor de Grado 3')
plt.axhline(0, color='gray', linewidth=0.5, ls='--')
plt.axvline(1, color='gray', linewidth=0.5, ls='--')  # Cambi√© el eje vertical a x = 1 para enfocar la expansi√≥n de Taylor
plt.ylim(-2, 2)  # Limitar el eje y para mejor visualizaci√≥n
plt.xlim(0, 2)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
